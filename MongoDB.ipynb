```python
# insert_one()
collection.insert_one(
                      {"name" : "user 1",
                       "age" : 20,
                       "gender" : "Male",
                       "email" : "user1@gmail.com"}
                       )
```

##Read Data :

```python
# find_one() :   return single document
# find()      :  return list of documents
```

```python
# collection.find_one({condition})
```

```python
collection.find_one({"name" : "user4"})
```

```python
collection.find_one({"name" : "user4"},
                    {"email" : 1, "_id" :  0})

## return only gender for given condition {"name" : "user4"}
```

```python
lst = collection.find({"gender" : "Male"})

list(lst)
```

```python
lst = collection.find({"gender" : "Male"},
                      {"name" : 1, "email" : 1, "_id" : 0})

list(lst)
```

```python
## Conditions :
$gt  --> greater than
$gte  --> greater than equal to
$eq   --> equal to
$ne   --> not equal to
$lt  --> less than
$lte  --> less than equal to
$in  --> from given list
```

```python
lst = collection.find({"age" : {"$lt" : 30}})

list(lst)
```

```python
lst = collection.find({"age" : {"$gte" : 30}})

list(lst)
```

```python
lst = collection.find({"age" : {"$gte" : 30}},
                      {"name" : 1, "email" : 1, "_id" : 0})

list(lst)
```

```python
lst = collection.find({"age" : {"$in" : [20, 30, 40, 50]}},
                      {"name" : 1, "email" : 1, "_id" : 0, "age": 1})

list(lst)
```

```python
## Combining Multiple conditions :
```

```python
$or   : if any one condition is true
$and  : both condition should be true
```

```python
lst = collection.find( {"$and" : [
    {"age" : {"$gt" : 30}},
    {"gender" : "Male"}
    ]
}
    )


list(lst)
```

```python
lst = collection.find({}, {"_id": 0})

# list(lst)
import pandas as pd

df = pd.DataFrame(list(lst))
```
